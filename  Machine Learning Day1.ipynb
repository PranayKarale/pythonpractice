# Plotting withn Numpy and matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(np.arange(10),np.arange(10))
```

# plot exponential


```python
plt.plot(np.arange(10),np.exp(np.arange(10)))
```

# Place the two graphs side by side

```python
plt.figure()
plt.subplot(211)
plt.plot(np.arange(10),np.arange(10))
plt.subplot(212)
plt.plot(np.arange(10),np.exp(np.arange(10)))
```

# Fours graphs side by side

```python
plt.figure()
plt.subplot(221)
plt.plot(np.arange(10),np.arange(10))
plt.subplot(222)
plt.scatter(np.arange(10),np.arange(10))
plt.subplot(223)
plt.plot(np.arange(10),np.exp(np.arange(10)))
plt.subplot(224)
plt.scatter(np.arange(10),np.exp(np.arange(10)))
```

```python
from sklearn.datasets import load_iris
iris = datasets.load_iris()
data = iris.data
target = iris.target
# Resize figure
plt.figure(figsize=(12,5))
#first subplot
plt.subplot(121)
#visualize first plot
plt.scatter(data[:,0],data[:,3],c = target)
#second subplot
plt.subplot(122)
#visualize second plot
plt.scatter(data[:,1],data[:,3],c = target)
```
